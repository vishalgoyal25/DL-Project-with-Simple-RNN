```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
### sentences
sent=[  'the glass of milk',
     'the glass of juice',
     'the cup of tea',
    'I am a good boy',
     'I am a good developer',
     'understand the meaning of words',
     'your videos are good',]
```

```python
sent
```

```python
## Define the vocabulary size
voc_size=10000
```

```python
### One Hot Representation
one_hot_repr=[one_hot(words,voc_size)for words in sent]
one_hot_repr
```

```python
## word Embedding Representation

from tensorflow.keras.layers import Embedding

from tensorflow.keras.utils import pad_sequences

from tensorflow.keras.models import Sequential
```

```python
import numpy as np
```

```python
sent_length = 8         # Maximum possible length of any Sentence ( Let's take 8)

embedded_docs = pad_sequences(one_hot_repr, padding='pre', maxlen=sent_length)

print(embedded_docs)
```

```python
## feature representation
dim=10
```

```python
import tensorflow as tf

model=Sequential()

model.add(tf.keras.Input(shape=(sent_length,)))

model.add(Embedding(voc_size, dim, mask_zero=True))    
#You're not passing data, Giving the structure to the model — telling it what kind of data to expect and how to process it.


## Output Shape---(8, 10) → 8 words each converted into 10-d embeddings
```

```python
model.compile(optimizer='adam', loss='mse')    ## Optimizer and Loss Function
```

```python
model.summary()
```

```python
model.predict(embedded_docs) 
# It will return the model’s output predictions for each input sample in embedded_docs

# Each word index is mapped to a 10 D vector
```
